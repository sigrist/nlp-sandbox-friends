```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
# Open some HTML data
with open('data/101pilot.htm', 'r') as file:
    data = file.read().replace('\n', '')
```

```python
# Create a soup instance with the HTML 
soup = BeautifulSoup(data, 'html.parser')
```

```python
print(soup.prettify())
```

### PadrÃ£o do texto

Alguns padrÃµes identificados:

* Todo conteÃºdo entre parenteses Ã© algo que nÃ£o Ã© da fala, e sim alguma explicaÃ§Ã£o do que estÃ¡ acontencendo
* O nome do interlocutor estÃ¡ entre as tags <B> e </B>
* MudanÃ§as de cena estÃ£o identificadas com [Scene: ......]

```python
# Now get all P tags
ps = soup.find_all('p')
```

```python
def parse_line(line):
    array = line.split(':')
    if (len(array) == 2):
        return array
    return ['','']
```

```python
lines = [parse_line(p.get_text()) for p in ps]
# O formato aqui vai ficar 
#     author: sentenc
```

```python
authors = [x[0] for x in lines]
sentences = [x[1] for x in lines]
df = pd.DataFrame({ 'author': authors, 'sentence': sentences})
```

```python
df.replace('', np.nan, inplace=True)
df = df.dropna()
```
